```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("matches (1).csv")
```

```python
data.head()
```

```python
data.shape
```

```python
# Top 5 players with most "Player of the Match" award
player_of_match=data["player_of_match"].value_counts()[:5]
player_of_match
```

```python
# Plotting Top 5 player with most Plyaer of the Match Award 
plt.figure(figsize=(8,5))
plt.bar(data["player_of_match"].value_counts()[:5].keys(),list(data["player_of_match"].value_counts()[:5]),color='y')
```

```python
# Teams with most number of wins
data["winner"].value_counts()
```

```python
# Teams with most number of toss wins
data["toss_winner"].value_counts()
```

```python
# Result type
data["result"].value_counts()
```

## Batting First

```python
# Team won batting first
batting_first= data[data["win_by_runs"]!=0]
batting_first.head()
```

```python
# Plotting a histogram for Win Margin by No.of matches batting first
plt.figure(figsize=(8,5))
plt.hist(batting_first["win_by_runs"])
plt.xlabel("Win By Runs")
plt.ylabel("No. of Matches")
```

```python
# Number of wins w.r.t each team batting first
list(batting_first["winner"].value_counts()[:3])
```

```python
# Number of wins w.r.t each team batting first
batting_first["winner"].value_counts()[:3]
```

```python
# Plotting a Bar chart for teams with most wins after batting first
plt.figure(figsize=(8,5))
plt.bar(batting_first["winner"].value_counts()[:3].keys(),list(batting_first["winner"].value_counts()[:3]),color=['blue','yellow','orange'])
plt.xlabel("Teams")
plt.ylabel("No. of Matches")
plt.show()
```

```python
# Plotting a pie chart for team winning most after batting first
plt.pie(list(batting_first["winner"].value_counts()[:10]),labels=batting_first["winner"].value_counts()[:10].keys(),autopct='%1.1f%%')
plt.show()
```

## Bowling First

```python
# Team won bowling first
bowling_first=data[data["win_by_wickets"]!=0]
bowling_first.head()
```

```python
# Plotting a histogram for Win Margin by No.of matches bowling first
plt.hist(bowling_first['win_by_wickets'])
plt.xlabel("Wins by wickets")
plt.ylabel("No. of matches")
plt.show()
```

```python
# Plotting a Bar chart for teams with most wins after bowling first
plt.bar(bowling_first["winner"].value_counts()[:3].keys(),list(bowling_first["winner"].value_counts()[:3]),color=['purple','blue','red'])
plt.xlabel("Teams")
plt.ylabel("No. of Matches")
plt.show()
```

```python
# Plotting a pie chart for team winning most after bowling first
plt.pie(list(bowling_first["winner"].value_counts()[:10]),labels=bowling_first["winner"].value_counts()[:10].keys(),autopct='%1.1f%%')
plt.show()
```
